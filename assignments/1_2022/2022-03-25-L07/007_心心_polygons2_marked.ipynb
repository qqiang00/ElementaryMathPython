```python
# 是不是正多边形的边数越多，内、外角和就越大？

# The greater the amount of sides, the greater the sum of interior angles will be, since 
# the interior angle itself is greater and there are more edges to multiply by.
# However, the exterior angles' sum will always be 360 degrees, since every exterior angle
# will be smaller with the number of edges being greater, but are multipled with higher numbers
# because there are more edges.

# Therefore when there are more sides in a polygon, the sum of interior angles will be greater,
# but the sum of exterior angles will not.
```
