### Problem 2

````python
# Solution 2

n_brain_total, n_feet_total = 35, 94
n_chicken, n_rabbit = 0, 0

while n_chicken <= 35:
    n_rabbit = 35 - n_chicken
    n_feet_current = 2 * n_chicken + 4 * n_rabbit
    if n_feet_current == 94:
        print(n_chicken, n_rabbit)
    
    n_chicken += 1
````

````python
# Solution 1

n_brain_total, n_feet_total = 35, 94
n_chicken, n_rabbit = 0, 0

while n_chicken < n_brain_total:
    n_rabbit = 0

    while n_rabbit < n_brain_total:
        if (n_chicken + n_rabbit == n_brain_total) and \
           (n_chicken * 2 + n_rabbit * 4 == n_feet_total):
            print("{} chicken and {} rabbit meet the requirement"\
                  .format(n_chicken, n_rabbit))
            
        n_rabbit += 1
    
    n_chicken += 1
````
