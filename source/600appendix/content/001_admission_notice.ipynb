## <center>《用Python学数学》第一期招生通知</center>

家长朋友们：

经过近一年的筹备和试讲，由叶强老师设计并主讲的《用Python学数学》初级编程培训班现在开始招收小学4年级（含）以上的小朋友们了。
如果您希望自己的孩子既能学习到一些编程知识和技能同时也能加深对相应数学知识的理解，那么这个课程值得您的关注。

### 基本特点

1. 本次招生主要针对北美的华人小朋友，但也招收来自中国大陆、香港、澳门和台湾地区的适龄小朋友。详见“入学条件”；
1. 本编程培训班从招生、授课、学期评估乃至行政管理等均参照北美地区小学的教学模式，内地的小朋友可以体验到不同的教学模式。详见“行政流程”；
1. 本次只开设初级班，时间为一学年，设三个学期，每个学期约16-18周，首学期开学日期约为2月中下旬。授课形式为：网上授课，小班教学。每周一次，每次约2个小时，中间休息15-20分钟。每周安排安排一次email答疑，答疑由助教组织，助教由有经验的高级学员或老师承担。参见“课程计划”；
1. 为保证教学质量，参加同一个网课学习的小朋友数量不超过6人，课堂上会有较多的一对一互动。详见“课堂形式”；
1. 本培训班不是免费的，但设立学费优惠政策和针对学生的激励措施。成绩相对较好的学生会得到实物奖励以及下一学期学费的折扣优惠，详见“学费、优惠和奖励计划”。

### 入学条件
1. 2021年以后出生（9周岁以上），小学4年级（含）以上的小、中学生，特别优秀的可适当放宽年龄、年级限制；
2. 主要面向北美地区小朋友，积极欢迎中国内地和港澳台地区的学生；
3. 对编程有较为浓厚的兴趣，有较好的数学基础更佳；
4. 语言和口头表达能力良好；
5. 有一定的打字速度。

### 行政流程
1. 有兴趣的学生家长自见到本通知开始即可填写报名表(另有提供）报名；
2. 老师收集报名表，对照入学条件录取或安排一次与学生和家长的视频面谈；
3. 老师会对每一个学生进行一次评估并排序。否满足入学条件以及报名先后影响排序结果。
4. 最终录取学生在2人以上即正常开班，排序在第6名以外的学生将不能被直接录取，需进入等候区；
5. 在开学后2周以内，如果有学生退出，等候区的学生依次会被通知入学；
6. 开学2周后仍然在等候区的学生将不再被录取；
7. 被录取的学生有资格注册学期，按照教学日历完成每一学期的学习；
8. 被录取有资格注册的学生在学期开学前一周仍未注册的将被视为自动放弃，
9. 因放弃注册空出的名额将会派法给等候区内排名最考前的学生；
8. 学生因病因事请假没有参加的课程没有补课，但会提供上课的完整视频供自学；
9. 学生因病因事请假没有参加期中和期末考试，协商日期安排一次补考，补考仍未参加则不得参加后续课程且不退还学费；
10. 学期结束考核及格可以进入下一学期学习；
11. 学期结束考核不及格不得进入下一学期学习，但有优先再下一年度入学的机会；
12. 期末的总体考核的分配为：平时作业占比为40%，期中考试占20%，期末考试占40%。

### 课程计划
1. 初级班暂定一学年共三个学期，时间跨度约为一个自然年左右。每个学期约为16-18周；
2. 目前暂定的可选的授课时间为（会有调整）：
   - 每周五加东晚上8:00-10:00，加西下午5:00-7:00, 北京时间周六上午9:00-11:00
   - 每周六加东晚上8:00-10:00，加西下午5:00-7:00, 北京时间周日上午9:00-11:00
2. 每周一次视频授课(暑假期间可能会略有不同），时间约为2个小时，中间安排1-2次休息，休息时间约为20分钟左右；
3. 每周针对每一个学生设立一次email答疑，答疑有助教负责，助教由由经验的高级学员或者老师承担；
3. 设立期中和期末考试周，考试采取视频监考形式。考试周不安排新的教学内容，但安排答疑；
5. 学期之间安排几周的假期；假期不安排任何形式的课程内容。

### 课堂形式
1. 视频授课以老师讲解和当场编程演示为主，课堂上安排一定的时间专门留给学生当场动手编程，并由老师进行一对一的指导；
2. 每次课后安排一定量的家庭作业，一般可在3-6小时内完成，难度较大的作业会有提示。学生可自由选择完成作业的时间；
3. 每次作业需要在下一次课前24小时交给老师批阅；
4. 学生在完成课后作业如果遇到困难，可以以邮件的形式向老师和助教求助，一般会在24小时内得到答复；
1. 每次课堂开始的时候会集中讲解作业中遇到的错误；
2. 每一个学期会有一次期中和期末考试，考试的形式为视频监考，考试内容为近期所学的知识，需要学生在有限的考试时间里独立编程解决；考试试题不安排讲解。

### 学费、优惠和奖励计划
1. 学费暂定为:CAD960(RMB4800)/学期，有效期为当年2022年整个学年。当加币兑换人民币在区间4.7-5.3:1，可选择任意一个币种支付；当汇率超出此范围，只能选择学员所在地常用的币种进行支付；
2. 重要时间节点：每个学期，已注册的学生最迟需在**开学后第二周截止前**完成缴费。如果发生逾期，将收取每日**1%的滞纳金**，超过**一周**仍未缴费，则自动丧失学籍，不得参加**第三周**的课程学习；
3. 与圈子相关的优惠：微信朋友圈好友享受**8折**优惠，朋友圈好友推荐的可享受**9折**优惠，优惠范围目前暂设为首个学年的三个学期。
4. 与成绩相关的优惠： 当班级人数超过（含）5人时，期末综合成绩排班级第一名的学生，下一个学期的学费可以再增加1折的优惠，即享受原价的7折（或8折）优惠；期末综合成绩排名第2的学生，相应增加的折扣为0.5折，即享受原价的7.5折（或8点5折）优惠。当班级人数为3-5人时，仅第一名享有0.5的折扣；当班级人数仅为2人时，不设优惠。优惠每个学期均根据上述条件设置，已经获得的优惠只有连续的下一个学期有效，持有优惠中断学习再恢复学习，优惠不再有效。
5. 奖励计划：对在学习中有出色表现的行为，考虑以奖学金或实物的形式额外进行奖励。一次奖励金额暂定为CAD20(RMB100)，单个学生一个学期最多能接受奖励3次。
6. 对确有经济困难但成绩特别优秀的学员，可在第二、三个学期与老师协商减免学费。
