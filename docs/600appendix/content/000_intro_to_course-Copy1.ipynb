## 附录一 初级课程介绍

每周90分钟授课，学习完成初级课程至少需要一个学年

### 0. 教学目的和计划
- 教授基本的编程知识和理念
- 从编程的角度加深对数学知识的理解
- 从编程的角度解决简单数学问题的能力
- 初步培养学生将复杂问题分解为多个简单问题的能力

### 1. 涉及的编程知识
- Python语言的基本语法
- 基本的输入输出
- 变量的声明和使用
- 常见的数据类型：整数型、字符串型、列表、元组
- 简单的逻辑运算：and, or, not
- 条件和循环语句
- 声明和实现方法，方法的参数和返回值的设计
- 库的概念以及turtle绘图库常用方法的使用
- 穷举算法

### 2. 涉及的数学知识
- 自然数、四则运算
- 比较数的大小，基本的逻辑运算
- 简单的几何图形（三角形、矩形、正方形、正多边形等）及其基本性质
- 简单几何图形的绘制，周长和面积的计算
- 认识角度、直角、锐角、钝角等
- 数轴、平面直角坐标系
- 平行于数轴的线段根据端点坐标计算线段长度
- 乘方的意义
- 和与差的平方的几何意义
- 毕达哥拉斯（勾股）定理及其集合意义
- 平面中任意两点间距离的计算
- 二次方根及其几何意义
- 轴对称和中心对称的数学表示形式
- 基本的统计概念：平均值、中位数等
- 绘制基本的统计图表
- 初步认识三角函数


### 3. 授课方式课课程计划
- 主讲：网课，每周1次，一次90分钟左右，中间休息15分钟，
- 答疑：email，每周一次，

### 4. 每次主讲时间安排

- 作业集中反馈：约15分钟
- 提出本讲问题：约15分钟
- 引导学生思考：约15分钟
- 现场编程演示：约30分钟
- 当次课程小结：约05分钟
- 解释课后作业：约10分钟

### 5. email答疑安排
- Email答疑每个学生一周回复一次，
- 建议学生在下一次上课前的2-3天前发邮件给老师，一般第二天会有回复。

### 6. 语言选择
- 口头交流：中文
- 讲义：中英文双语
- 书面作业：英文或中文（学生自主选择）

### 7. 家长配合
- 提醒孩子提前10分钟开始准备上课电脑，协助或让孩子自己打开编程环境
- 课后督促孩子们完成布置的编程作业
- 提醒孩子提交作业
- 留意孩子作业的评语

### 8. 第一次课程前准备
- 帮孩子搭建编程环境
- 留意孩子电脑的操作系统MacOS, Windows
- Jupyter Lab(Notebook)介绍
