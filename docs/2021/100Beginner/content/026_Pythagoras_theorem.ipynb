```python
#1. æ ‡è®°ä¸‰ä¸ªç‚¹:A(10, 9), B(18, 9), C(10, 15)ï¼Œä¸‰è§’å½¢ABCæ˜¯ä¸€ä¸ªç›´è§’ä¸‰è§’å½¢ï¼Œ
#   ç”¨é»‘è‰²ç»˜åˆ¶è¿™ä¸ªä¸‰è§’å½¢ï¼›æ ‡è®°ä¸ºTri0; 
#   å°†è¿™ä¸ªç›´è§’ä¸‰è§’å½¢çš„ä¸‰æ¡è¾¹ä»çŸ­åˆ°é•¿åˆ†åˆ«æ ‡è®°ä¸º ğ‘ , ğ‘ å’Œ ğ‘ ;
    
a, b = 6, 8      # two side lengths of the right triangle ABC

A = (10, 9)              # Start Point, vertex of triangle ABC with right angle
B = (A[0] + b, A[1])
C = (A[0], A[1] + a)
Tri0 = [B, A, C]
polygon(Tri0, color="black", center_text="Tri0",
        side_texts="bac", 
        side_text_font = ("Arial", 16, "italic"),
        side_text_offsets=[(-0.5, -1.5), (-0.5, -0.5), (0.5, 0)],
        side_text_color="red",
        marker_texts = ["B"+str(B), "A"+str(A), "C"+str(C)])
```

```python
# 2. ä»¥ç›´è§’ä¸‰è§’å½¢çš„ç›´è§’è¾¹ABä¸ºè¾¹é•¿ï¼Œåœ¨ä¸‰è§’å½¢ABCçš„ä¸‹æ–¹å»ºç«‹ä¸€ä¸ªæ­£æ–¹å½¢ï¼Œä½¿å¾—ä¸‰è§’å½¢çš„è¾¹ABä¸ç»˜åˆ¶çš„
#    æ­£æ–¹å½¢é‡åˆï¼›æ„æˆè¿™ä¸ªæ­£æ–¹å½¢çš„å¦å¤–ä¸¤ä¸ªé¡¶ç‚¹å‘½åä¸ºDå’ŒEï¼Œå…¶ä¸­è¾¹DEä¸è¾¹ACå¹³è¡Œï¼ŒDä½äºEçš„å·¦ä¾§ï¼›
#    è¯¥æ­£æ–¹å½¢åˆ™ç§°ä¸ºABEDã€‚æ ‡è®°ç‚¹Då’ŒEï¼Œå¹¶ç”¨çº¢è‰²ç»˜åˆ¶è¿™ä¸ªæ­£æ–¹å½¢çš„è¾¹ï¼Œæ ‡è®°ä¸ºSqr1ï¼›
D = (A[0], A[1] - b)
E = (B[0], B[1] - b)
Sqr1 = [A, B, E, D]
polygon(Sqr1, color="green", center_text="Sqr1")
mark(D, "D"+str(D))
mark(E, "E"+str(E))
```

```python
# 3. ä»¥ç›´è§’ä¸‰è§’å½¢çš„ç›´è§’è¾¹ACä¸ºè¾¹é•¿ï¼Œåœ¨ä¸‰è§’å½¢ABCçš„å·¦ä¾§å»ºç«‹ä¸€ä¸ªæ­£æ–¹å½¢ï¼Œä½¿å¾—ä¸‰è§’å½¢çš„è¾¹ACä¸ç»˜åˆ¶çš„
#    æ­£æ–¹å½¢é‡åˆï¼›æ„æˆè¿™ä¸ªæ­£æ–¹å½¢çš„å¦å¤–ä¸¤ä¸ªé¡¶ç‚¹å‘½åä¸ºFå’ŒGï¼Œå…¶ä¸­è¾¹GFä¸è¾¹ACå¹³è¡Œï¼ŒFä½äºGçš„ä¸Šæ–¹ï¼›
#    æ­£æ–¹å½¢åˆ™ç§°ä¸ºACFGã€‚ç”¨çº¢è‰²ç»˜åˆ¶è¿™ä¸ªæ­£æ–¹å½¢çš„è¾¹ï¼›æ ‡è®°ä¸ºSqr2
F = (C[0]-a, C[1])
G = (A[0]-a, A[1])
Sqr2 = [A, C, F, G]
polygon(Sqr2, color="red", center_text="Sqr2")
mark(F, "F"+str(F))
mark(G, "G"+str(G))
```

```python
#  4. æ ‡è®°ä¸¤ä¸ªç‚¹H(24, 17), I(16, 23)ï¼Œå››è¾¹å½¢BHICæ˜¯ä¸€ä¸ªè¾ƒå¤§çš„æ­£æ–¹å½¢ï¼›
#     ç”¨é’è‰²ç»˜åˆ¶è¿™ä¸ªæ­£æ–¹å½¢ï¼Œæ ‡è®°ä¸ºSqr3
H = (24, 17)                        # (B[0] + 10*3/5, B[1] + 10*4/5)
I = (16, 23)                        # (C[0] + 10*3/5, C[1] + 10*4/5)
Sqr3 = [B, H, I, C]
polygon(Sqr3, color="cyan", center_text="Sqr3")
mark(H, "H"+str(H))
mark(I, "I"+str(I))
```
