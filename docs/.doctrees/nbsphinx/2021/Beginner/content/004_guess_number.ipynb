```python
from random import randint, random
num = randint(1, 10)
print("I (the computer) has generated a number between 1 and 10")
print("Now it's your turn to guess what the number is:")
guess_qiang = int(input("Qiang's guess: "))
guess_tony =  int(input("Tony's guess: "))
guess_jason = int(input("Jason's guess: "))
guess_sophie = int(input("Sophie's guess: "))

if num == guess_qiang:
    print("Qiang win")
elif num == guess_tony:
    print("Tony win")
elif num == guess_jason:
    print("Jason win")
elif num == guess_sophie:
    print("Sophie win")
else:
    print("No body win")
    
print("the number generated by computer is: ", num)
```
